```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots as sp
plt.style.use(['science','notebook','grid'])
```

```python
lista_m = list()
lista_f = list()

with open('masc.txt','r',encoding='utf-8') as arquivo_m:
    lista_m = [linha.replace('\n','') for linha in arquivo_m]
    lista_m = [lista_m[i].split('\t') for i in range(len(lista_m))]

x_masc = np.asarray([lista_m[i][0] for i in range(len(lista_m)-1)],dtype=float)
y_masc = np.asarray([lista_m[i][1] for i in range(len(lista_m)-1)],dtype=float)

with open('fem.txt','r',encoding='utf-8') as arquivo_f:
    lista_f = [linha.replace('\n','') for linha in arquivo_f]
    lista_f = [lista_f[i].split('\t') for i in range(len(lista_f))]

x_fem = np.array([lista_f[i][0] for i in range(len(lista_f)-1)]).astype(float)
y_fem = np.array([lista_f[i][1] for i in range(len(lista_f)-1)]).astype(float)

plt.figure(figsize=(6,4))
plt.plot(x_masc,y_masc,'o',color='r', label='Dados Masc.')
plt.plot(x_fem,y_fem,'o',color='g', label='Dados Fem.')
plt.xlabel('Peso (kg)')
plt.ylabel('Altura (cm)')
plt.legend(fontsize=10)
plt.show()
```

```python
fig,ax = plt.subplots(1,1,figsize=(6,4))
ax.plot(x_masc,y_masc,'o',color='r')
ax.plot(x_fem,y_fem,'o',color='g')
ax.set_xlabel('Peso (kg)')
ax.set_ylabel('Altura (cm)')
ax.set_title('Altura x Peso (cm x kg)')
#ax.text(0,1,'MacFem',transform=ax.transformAxes)
plt.show()
```

```python
fig,ax = plt.subplots(1,1,figsize=(6,4))
ax.plot(x_masc,y_masc,'o',color='r',label='Masc')
ax.plot(x_fem,y_fem,'o',color='g',label='Fem')
ax.set_xlabel('Peso (kg)\n')
ax.set_ylabel('Altura (cm)')
ax.set_title('Altura x Peso (cm x kg)')
#ax.text(0.1,0.7,'MacFem',transform=ax.transAxes)
ax.tick_params(axis='both',labelsize=10)
ax.legend()
plt.show()
```

```python
textstr1 = r'$\Sigma$ = Masculino'

fig,axes = plt.subplots(1,2,figsize=(14,6))

ax_1 = axes[0]
ax_1.plot(x_masc,y_masc,'o',color='r')
#ax.plot(x_fem,y_fem,'o',color='g')
ax_1.set_xlabel('Peso (kg)\n')
ax_1.set_ylabel('Altura (cm)')
#ax_1.set_title('Altura x Peso (cm x kg)')
ax_1.text(0.1,1.5,textstr1,transform=ax.transAxes,bbox=dict(facecolor='white',edgecolor='black'))
ax_1.tick_params(axis='both',labelsize=10)

ax_2 = axes[1]
#ax_1.plot(x_masc,y_masc,'o',color='r')
ax_2.plot(x_fem,y_fem,'o',color='g')
ax_2.set_xlabel('Peso (kg)\n')
ax_2.set_ylabel('Altura (cm)')
#ax_2.set_title('Altura x Peso (cm x kg)')
ax_2.text(1.5,1.5,'MacFem',transform=ax.transAxes,bbox=dict(facecolor='white',edgecolor='black'))
ax_2.tick_params(axis='both',labelsize=10)

plt.show()
```

```python
fig, ax = plt.subplots(1,3,figsize=(12,4))
ax[0].hist(x_masc,bins=10,label='Pesos Masculinos',color='b',edgecolor='black')
ax[0].legend(fontsize=7)
ax[0].set_xticks(np.arange(int(np.min(x_masc)),int(np.max(x_masc))+2,2))
ax[0].set_yticks(np.arange(3,22,3))
ax[0].tick_params(axis='both',labelsize=10)
ax[0].set_xlim([63, 76])
ax[0].set_ylim([0, 21])
ax[0].set_xlabel('Pesos kg',fontsize=7)
ax[0].set_ylabel('Frequência (pessoas)',fontsize=7)

ax[1].hist(x_fem,bins=10,label='Pesos Femininos',color='r',edgecolor='black')
ax[1].legend(fontsize=7)
ax[1].set_xticks(np.arange(int(np.min(x_fem)),int(np.max(x_fem)+2),2))
ax[1].set_yticks(np.arange(3,22,3))
ax[1].tick_params(axis='both',labelsize=10)
ax[1].set_xlim([56, 71])
ax[1].set_ylim([0, 21])
ax[1].set_xlabel('Pesos kg',fontsize=7)
ax[1].set_ylabel('Frequência (pessoas)',fontsize=7)

ax[2].bar('Masc', height=[np.mean(x_masc)], width=0.5, color='b', edgecolor='black',label='Média de Pesos Masc')
ax[2].bar('Fem', height=[np.mean(x_fem)],label='Média de Pesos Fem', width=0.5, color='r', edgecolor='black')
ax[2].legend(fontsize=8)
ax[2].tick_params(labelsize=10)
ax[2].set_ylim([0, 80])
ax[2].set_xlabel('Sexo',fontsize=7)
ax[2].set_ylabel('Média (kg/pessoas)',fontsize=7)

plt.show()
```
